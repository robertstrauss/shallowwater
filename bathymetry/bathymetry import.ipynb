```python
h = bath.variables['elevation'][0:1000, 0:1000] # crashes on unlimited size! limit to small
h
```

```python
x=np.linspace(0, h.shape[0]-1, h.shape[0])
y=np.linspace(0, h.shape[1]-1, h.shape[1])
xx,yy=np.meshgrid(x, y)
fig=plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(xx, yy, h, rstride=20, cstride=20)
```
